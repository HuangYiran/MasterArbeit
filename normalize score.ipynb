进行规范化的原因有：
- 并没有取所有的数据，所以数据的mean并不是0
- 并不是把这些数据作为输入，而是作为拟合的对象，没了更好的进行拟合，所以进行规范化。

这个文件尝试把分数的值域：规范到[-1,1]<br>
有R可以知道分数的分布情况：<br>
 Min.   :-4.17367  <br>
 1st Qu.:-0.83528  <br>
 Median : 0.01462  <br>
 Mean   :-0.08855  <br>
 3rd Qu.: 0.75517  <br>
 Max.   : 1.96235  <br>
 我们这里取最大最小值分别为3， -1

```python
dir_score = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/data/data_scores"
dir_tgt = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/data/normalized_data_scores"
```

```python
f_in = open(dir_score, "r")
f_out = open(dir_tgt, "w")
```

```python
max = 3
min = -3
new_max = 1
new_min = -1
```

```python
for line in f_in:
    score = float(line.strip())
    #print(score)
    nor_score = (score-min)*(new_max - new_min)/(max-min) + new_min
    f_out.write(str(nor_score)+"\n")
```

```python
f_in.close()
f_out.close()
```
