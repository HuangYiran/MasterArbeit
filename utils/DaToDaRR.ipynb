```python
def horToVer(li, threshold = 0.1, index = 0):
    length = len(li)
    tmp = []
    for i in range(length):
        j = i + 1
        while j < length:
            if float(li[i]) - float(li[j]) > threshold:
                rr = 1
            elif float(li[i]) - float(li[j]) < -threshold:
                rr = -1
            else:
                rr = 0
            tmp.append(str(index) + "," + li[i] + "," + li[j] + "," + str(rr))
            j += 1
    return tmp
```

```python
# get the original da data
#ref = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/plan_a_source_en.de/extracted_data/data_ref"
src = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/plan_a_source_en.de/extracted_data/data_source"
#sys = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/plan_a_source_en.de/extracted_data/data_sys_out"
scores = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/plan_a_source_en.de/extracted_data/data_scores"
dir_darr = "/Users/ihuangyiran/Documents/Workplace_Python/data/MasterArbeit/plan_a_source_en.de/extracted_data/darr"
```

```python
# load the data
data_src = [li.rstrip('\n') for li in open(src)]
data_scores = [li.rstrip('\n') for li in open(scores)]
```

```python
# set attributes
darr = [] # list of string, sid, s1, s2, rr
sid = 1
num_repeat = 1 # how many time is the sentence repeated
current_sent = data_src.pop(0)
scores_batch = [] # store the scores of the same sentence
scores_batch.append(data_scores.pop(0))
```

```python
for (src, score) in zip(data_src, data_scores):
    if src == current_sent:
        num_repeat += 1
    else:
        if num_repeat > 1:
            darr.extend(horToVer(scores_batch, index = sid))
        num_repeat = 1
        current_sent = src
        scores_batch = []
        sid += 1
    scores_batch.append(score)
print darr
```
